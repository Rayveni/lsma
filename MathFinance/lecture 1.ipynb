# Деривативы(Производные фин. инструменты)
В математике: $ \frac{df}{dx}=\lim\limits_ {\Delta x \rightarrow 0} \frac {f(x+\Delta x)-f(x)}{\Delta x}$ <br>
В финансах функция , к примеру, может представлять значение портфеля(в $) , а агрументом функции-**"риск фактор"**.
Пример риск фактора - процентные ставки, цена акций, цена на сырую нефть или природный газ, обмен валюты.
<br>
Соглашение о покупке в будущем определенного количества актива по цене, согласованной сегодня, является примером производного финансового инструмента.

Подобное  соглашение, подписанное в прошлом (сегодня или раньше) обладает определенной ценностью сегодня.(насколько убыточной/выигрышной будет сделка на дату поставки товара).

Общий подход в оценке заключается в расчете рыночной стоимости:<br>
$V=Qty∙(Fwd-K)∙Df$,где:<br>
 **Qty** - количество актива<br>
**Fwd** - форвардная цена сегодня <br>
**К** - договорная фиксированная цена<br>
**Df** - коэффициент дисконтирования; при текущих безрисковых инвестиционных ставках(например депозит в банке/курс ЦБ) ,показывает сколько нужно инвестировать сегодня, чтобы иметь 1 доллар на срок погашения. 


Интерпретация:
**Оценка рыночной стоимости** показывает прибыль или убыток, которые получил бы трейдер, если бы он закрыл существующую позицию на текущем уровне рынка.

**Дериватив** - это финансовый инструмент, стоимость которого определяется на основе базового актива или группы активов в более поздний момент времени (срок погашения или истечение срока).<br>
Дериватив определяется в договоре между двумя или более сторонами. Стоимость дериватива до его погашения оценивается на форвардных / фьючерсных рынках.<br>
Наиболее распространенные базовые активы включают акции, облигации, товары, валюты, процентные ставки и рыночные индексы.

# Базовые инструменты

Финансовые инструменты, используемые в качестве входных данных при расчётах, называются **базовыми инструментами**.

Форвардные цены, процентные ставки, курсы валют, а иногда и цены на акции являются **базовыми инструментами**.
Их значения (рыночные уровни) изменяются между датой транзакции и сроком погашения. Предполагается, что они часто следуют определенным случайным процессам.
